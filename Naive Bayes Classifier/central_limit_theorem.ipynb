```python
import math
from turtle import distance
import numpy as np
import random

num_of_samps = 10000
sample_size = 30

means = []
for _ in range(num_of_samps):
    mean = 0
    for i in range(sample_size):
        x = random.randint(0, 500)  
        y = random.randint(0, 500)
        
        distance = x+y
        
        mean += distance
        
    mean /= sample_size
    
    means.append(mean)
    
    
import seaborn as sns
sns.set_theme()

sns.distplot(means)

```
